### Section 1

#### 1. Which companies applied for the largest number of H1B visas where the job opening was located in NYC? Please describe any issues you may encounter summarizing the data by employer name.

```python
h1b_NewYork = h1b_data[h1b_data['LCA_CASE_EMPLOYER_CITY'] == 'NEW YORK'] #new dataframe - NY data 
h1b_NewYork
```

```python
#companies applied for the largest number of H1B in NYC 
h1b_NewYork['count'] = 1
employer_count = h1b_NewYork[['LCA_CASE_EMPLOYER_NAME','count']]
top = employer_count.groupby('LCA_CASE_EMPLOYER_NAME').count().sort(columns='count', ascending=False).head(20)
print ('Top NYC employers who sponsor H1B:')
print (' ')
for e, c in zip(top.index, top['count']):
    print ('{0} | {1}'.format(e, c))
```

#### 2. Calculate the mean and standard deviation of wages proposed for workers located in New York City and Mountain View. Are the average wages in these two locations statistically different? What factors could explain the results?


```python
h1b_NewYork.columns
```

```python
h1b_NewYork['LCA_CASE_WAGE_RATE_FROM'].mean() #calculation of mean for NY
```

```python
h1b_NewYork['LCA_CASE_WAGE_RATE_FROM'].std() #calculation of standard deviation for NY
```

```python
h1b_MountainView = h1b_data[h1b_data['LCA_CASE_EMPLOYER_CITY'] == 'MOUNTAIN VIEW'] #Mountain View data
h1b_MountainView
```

```python
h1b_MountainView['LCA_CASE_WAGE_RATE_FROM'].mean()#calculation of mean for MV
```

```python
h1b_MountainView['LCA_CASE_WAGE_RATE_FROM'].std()#calculation of standard deviation for MV
```

```python
NY_mean = h1b_NewYork['LCA_CASE_WAGE_RATE_FROM'].mean()
print ('NY mean:', NY_mean)
MV_mean = h1b_MountainView['LCA_CASE_WAGE_RATE_FROM'].mean()
print ('MV mean:', MV_mean)
```

```python
NY_cases = h1b_NewYork['LCA_CASE_NUMBER'].count()
MV_cases = h1b_MountainView['LCA_CASE_NUMBER'].count()
print ('NY cases:', NY_cases)
print ('MV cases:', MV_cases)
```

The average wages for NY and MV are different, but number of cases in NY is almost 5 times biger than number of cases in MV. If we look into results of standard deviation for proposed wages for both locations, it is clear that NY wages is widely distributed about the mean, while MV wages are clustered about the mean.

```python
#Decision period
```

```python
h1b_data['decision_period'] = (h1b_data['DECISION_DATE'] - h1b_data['LCA_CASE_SUBMIT']).dt.days
```

```python
dec_per_max = h1b_data['decision_period'].max()
dec_per_min = h1b_data['decision_period'].min()
dec_mean = h1b_data['decision_period'].mean()
dec_std = h1b_data['decision_period'].std()
print ('Max days for decision:', dec_per_max)
print ('Min days for decison:', dec_per_min)
print ('Average days for decision:', dec_mean)
print ('Standard  deviation:', dec_std)
```
